```python
import importlib
import subprocess
import sys

def packageCheck(path: str="requirements.txt") -> None:
    """Installs packages from requirements.txt, but avoids reinstalling if they're already installed."""

    with open(path, "r") as f:
        packages = [line.strip() for line in f.readlines() if line.strip() and not line.strip().startswith("#")]

    for package in packages:
        try:
            importlib.import_module(package)
        except ImportError:
            print(f"{package} is not installed. Installing {package}...")
            subprocess.check_call([sys.executable, "-m", "pip", "install", package])
            print(f"{package} has been installed.")
```

```python
# %pip list
packageCheck("./requirements.txt")
```
